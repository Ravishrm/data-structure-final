Que 4. Describe how dictionary store data?

Ans. Dictionary can store data in the form of key values pair.allowing fast and efficient access to values using unique keys.
In Python, a dictionary is an unordered, mutable collection of key-value pairs. Internally, Python dictionaries use a hash table to store and retrieve data efficiently.
Each key in the dictionary is passed through a hash function (using the built-in hash() function).

This produces a hash value, which is an integer.The hash value is used to compute an index where the key-value pair is stored.The actual key and value are stored together in the slot to allow lookup, since different keys can have the same hash As more elements are added, the dictionary automatically resizes and rehashes to maintain performance.
